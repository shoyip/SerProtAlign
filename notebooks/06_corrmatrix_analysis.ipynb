```python
import numpy as np
import matplotlib.pyplot as plt
from alntk.plotting import default_plot_style
color_cycle = default_plot_style()
```

```python
ctilde = np.genfromtxt('../../corrmatrix/ctilde.csv', delimiter=',', dtype=float)
ctilde_iter = np.genfromtxt('../../corrmatrix/ctilde_iter.csv', delimiter=',', dtype=float)
```

```python
plt.figure()
plt.hist(ctilde.flatten(), bins=100, label='Full Alignment', edgecolor='#'+color_cycle[0], facecolor=(0,0,0,0))
plt.hist(ctilde_iter.flatten(), bins=100, label='Truncated Alignment', edgecolor='#'+color_cycle[1], facecolor=(0,0,0,0))
# plt.ylim([0, 50000])
# plt.xlim([0, 0.15])
plt.yscale('log')
plt.legend()
plt.show()
```
